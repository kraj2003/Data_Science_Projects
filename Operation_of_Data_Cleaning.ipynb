```python
#calculate the mean valueof the  normalised_losses
avg_nor_loss=df1['Normalized-losses'].astype("float").mean(axis=0)
print("avg mean",avg_nor_loss)

# avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
# print("Average of normalized-losses:", avg_norm_loss)
```

```python
#replace the null value with the mean of normalized-losses
df1['Normalized-losses'].replace(np.NaN,avg_nor_loss,inplace=True)
df1['Normalized-losses'].isnull()
```

```python
#same thing with 'bore','stoke','horsepower','Peak-rmp'
avg_bore=df1['Bore'].astype("float").mean(axis=0)
avg_stoke=df1['Stroke'].astype("float").mean(axis=0)
avg_horsepower=df1['Horsepower'].astype("float").mean(axis=0)
avg_peak=df1['Peak-rpm'].astype("float").mean(axis=0)
df1['Bore'].replace(np.NaN,avg_bore,inplace=True)
df1['Stroke'].replace(np.NaN,avg_stoke,inplace=True)
df1['Horsepower'].replace(np.NaN,avg_horsepower,inplace=True)
df1['Peak-rpm'].replace(np.NaN,avg_peak,inplace=True)
df1.isnull().sum()
```

```python
df1['Num-of-doors'].value_counts()
```

```python
df1['Num-of-doors'].value_counts().idxmax()
```

```python
df1['Num-of-doors'].replace(np.NaN,'Four',inplace=True)
df1['Num-of-doors'].isnull().sum
```

```python
#correcting datatypes
df1.dtypes
```

```python
df1[['Peak-rpm','Bore','Stroke','Price']]=df1[['Peak-rpm','Bore','Stroke','Price']].astype("float")
```

```python
#normalisation
df1['Length']=df1['Length']/df1['Length'].max()
df1['Width']=df1['Width']/df1['Width'].max()
df1['Height']=df1['Height']/df1['Height'].max()
df1[['Length','Width','Height']].head()
```

```python
#creating indicator variable or dummy variable
# we use regression in modeling and regresion doesn't understand words only numbers
# here we will convert fuel-type into indicator variable. as fuel-type contains values as either gas or diesel ,
#so we will covert it into 1's and 0's
dummy_variable1=pd.get_dummies(df1['Fuel-Type'])
dummy_variable1.head()
```

```python
#merge the two dataframes
df1=pd.concat([df1,dummy_variable1],axis=1)
#dropthe fuel-type coumn
df1.drop('Fuel-Type',axis=1,inplace=True)
df1.head()
```

```python
#same indicator variable on "aspiraton " column
#first counts the total number of values in the column
df1['Aspiration'].value_counts()
```

```python
dummy_variable2.rename(columns={'std':'standard',"turbo":"Turbo"},inplace=True)
dummy_variable2.head()
```

```python
#merge the dataframes
df1=pd.concat([df1,dummy_variable2],axis=1)
#drop the as[iration column
df1.drop('Aspiration',axis=1,inplace=True)
df1.head()
```

## basics of grouping

```python
df['Drive-wheels'].unique()
```

```python
df_group1=df1[['Drive-wheels','Body-style','Price']]
```

```python
df_group1=df_group1.groupby(['Drive-wheels'],as_index=False).mean()
df_group1
```

```python
group=df1[['Drive-wheels','Body-style','Price']]
df_group2=group.groupby(['Drive-wheels','Body-style'],as_index=False).mean()
df_group2
```

```python
df_grouped=df_group2.pivot(index='Drive-wheels',columns='Body-style')
df_grouped
```

```python
df_grouped=df_grouped.fillna(0)
df_grouped
```

```python
# same goes with body-style and price
pgroup=df1[['Body-style','Price']]
pgroup=pgroup.groupby(['Body-style'],as_index=False).mean()
pgroup
```
