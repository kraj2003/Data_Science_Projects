```python
# importing the libraries and reading house_prices.csv
import pandas as pd
import numpy as np
file=pd.read_csv("datasets/house_prices.csv")
file.head(3)
```

```python
file.shape
```

```python
file.isnull().sum()
```

```python
# droping columns that has many null values and the one which are not to be used
file.drop(columns=["Dimensions","Plot Area","Super Area","Society","Car Parking",'Price (in rupees)','Transaction','facing',"overlooking","Index"],inplace=True)
file.head()
```

```python
file.duplicated().sum()
```

```python
# droping the rows that has null values in it
file.dropna(inplace=True)
file.head()
file.isnull().sum()
```

```python
file.shape
```

```python
# sorting by house prices
house_prices=file.sort_values(by="Amount(in rupees)", ascending=True)
house_prices.head(5)
```

```python
# houses that are in goa 
goa=house_prices.loc[house_prices['location']=='goa']
goa.head(3)
```

```python
# filtering out the furnished houses
furnished=goa.loc[house_prices['Furnishing']=='Furnished']
furnished.head(3)
```

```python
# filter on ownership
ownership=furnished.loc[furnished['Ownership'].isin(['Freehold','Co-operative Society'])]
ownership.head(3)
# furnished['Ownership'].unique()
```

```python
# filter on status
status=ownership.loc[ownership['Status']=='Ready to Move']
status.head(3)
```

```python
status['Status'].unique()
```

```python
# creating the final dataframe
final_dataframe=pd.DataFrame(status)
final_dataframe
```

```python
final_dataframe=final_dataframe.reset_index()
# final_dataframe=final_dataframe.drop(["index","level_0"])
final_dataframe.head(2)
# print(final_dataframe.info())
```
